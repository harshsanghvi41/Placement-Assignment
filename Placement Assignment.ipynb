## 1. Create a function in python to read the text file and replace specific content of the file.

```python
f = open('example.txt','r+')
f.write("This is a placement assignment")
f.close()

def file_op():
    
    try:
        f = open('example.txt','r')
        data = f.read()
        data = data.replace('placement', 'screening')
        f.close()
        f = open('example.txt','r+')
        f.write(data)
        f.close()
        
    finally:
        f = open('example.txt','r')
        for i in f:
            print(i,end='')
        f.close()
    
    
file_op()
```

## 2. Demonstrate use of abstract class, multiple inheritance and decorator in python using examples.

### Multiple Inheritance

```python
class first_inheritance:
    
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
        
    def test(self):
        print('this is a method of first_inheritance class')
        
class second_inheritance(first_inheritance): 
    def test1(self):
        print('this is a method from second_inheritance class')
        
class multiple_inheritance(second_inheritance):
    def test2(self):
        print('this is a method of multiple_inheritance class')
```

```python
con = multiple_inheritance('FSDS','ML','Ai')
```

### Decorator

```python
class deco:
    def uppercase_decorator(function):
        def wrapper():
            func = function()
            make_uppercase = func.upper()
            return make_uppercase

        return wrapper
    
    @uppercase_decorator
    def say_hi():
        return 'full stack data science'
    
d = deco
d.say_hi()
```

### Abstract class

```python
class Animal():

    def move(self):
        pass

class Human(Animal):

    def move(self):
        print("I can walk and run")

class Snake(Animal):

    def move(self):
        print("I can crawl")


R = Human()
R.move()

K = Snake()
K.move()
```
